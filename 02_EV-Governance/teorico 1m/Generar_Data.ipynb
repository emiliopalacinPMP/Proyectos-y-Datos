```python
import pandas as pd
import numpy as np

# --- PARTE 1: LÍNEA BASE (WBS) ---
wbs_data = [
    ["1.0", "Ingeniería", 100000], ["1.1", "Estudios de Suelo", 20000], ["1.2", "Diseño Eléctrico", 50000], ["1.3", "Planos As-Built", 30000],
    ["2.0", "Procura", 450000], ["2.1", "Transformador Potencia", 300000], ["2.2", "Celdas de Media Tensión", 100000], ["2.3", "Cables y Conectores", 50000],
    ["3.0", "Obras Civiles", 200000], ["3.1", "Excavación y Cimentación", 80000], ["3.2", "Malla a Tierra", 40000], ["3.3", "Edificio de Control", 80000],
    ["4.0", "Montaje y Pruebas", 250000], ["4.1", "Montaje Electromecánico", 150000], ["4.2", "Pruebas FAT/SAT", 60000], ["4.3", "Puesta en Servicio", 40000]
]
# Expandir a 100 actividades para simulación
df_linea_base = pd.DataFrame(wbs_data, columns=['WBS', 'Actividad', 'BAC'])
df_linea_base.to_csv('linea_base.csv', index=False)

# --- PARTE 2: DATOS DE CAMPO (SPI 0.81, CPI 0.82) ---
n_registros = 2100
spi, cpi = 0.81, 0.82
# PV Planificado a la fecha: $800,000
pv_actual = 800000
ev_objetivo = pv_actual * spi # $648,000
ac_objetivo = ev_objetivo / cpi # $790,243

registros = []
for i in range(n_registros):
    id_act = np.random.choice(df_linea_base['WBS'])
    costo_base = (ev_objetivo / n_registros)
    registros.append({
        'ID_Registro': i+1,
        'WBS': id_act,
        'EV': costo_base,
        'AC': costo_base / cpi * np.random.uniform(0.98, 1.02),
        'Fecha': '2025-10-15'
    })

df_campo = pd.DataFrame(registros)
df_campo.to_csv('datos_campo.csv', index=False)
```
